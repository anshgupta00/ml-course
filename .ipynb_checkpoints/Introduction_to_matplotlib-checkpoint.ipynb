```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]  # Define x values
y = [50, 100, 200, 250]  # Define y values

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
# %matplotlib inline
# import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[6,7,8,9,4]
fig,ax=plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
fig.savefig("images/sample-plot.png")
plt.show();


```
