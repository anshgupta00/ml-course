```python
# %matplotlib inline
# import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[6,7,8,9,4]
fig,ax=plt.subplots(figsize=(5,5))
ax.plot(x,y)
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
fig.savefig("images/sample-plot.png")
plt.show();


```

```python
import pandas as pd
import numpy as np
```

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales
```

```python
In [2]: ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))

In [3]: ts = ts.cumsum()

In [4]: ts.plot()
# Out[4]: <matplotlib.axes.AxesSubplot at 0x5dfb2d0>
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','', regex=True)
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sales Date"]=pd.date_range("1/1/2024", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sales Date', y='Total Sales')
```
